```python
import csv
frbcat_file = '../frbcat_20191016.csv'

#FRB_dtype = [('ID','S'),('DM','f'),('DM_gal','f'), ('RM','f'),('tau','f'),('host_redshift','S'), ('tele','S')]
FRB_dtype = [('ID','S9'),('DM','f'),('DM_gal','f'), ('RM','S10'),('tau','S10'),('host_redshift','S4'), ('tele','S10')]



def GetFRBcat( telescope=None, RM=None, tau=None ):
    FRBs = []
    ii = 0
    with open( frbcat_file, 'rb') as f:
        reader = csv.reader( f )
        header = np.array(reader.next())
        i_ID = 0
        i_DM = np.where( header == 'rmp_dm' )[0][0]
        i_DM_gal = np.where( header == 'rop_mw_dm_limit' )[0][0]
        i_RM = np.where( header == 'rmp_rm' )[0][0]
        i_tau = np.where( header == 'rmp_scattering' )[0][0]
        i_zs = np.where( header == 'rmp_redshift_host' )[0][0]
        i_tele = np.where( header == 'telescope' )[0][0]
        i_s = [i_ID, i_DM, i_DM_gal, i_RM, i_tau, i_zs, i_tele]  ## order must fit order of FRB_dtype
        for row in reader:
            if telescope and ( row[i_tele] != telescope ) :
                continue
            if tau and ( row[i_tau] == 'null' ) :
                continue
            if RM and ( row[i_RM] == 'null' ) :
                continue
            FRBs.append( tuple( [ row[i].split('&')[0] for i in i_s ] ) )
            ii += 1
    print "N = %i" % ii
    return np.array( FRBs, dtype=FRB_dtype )
```

Here we estimate the redshift based on DM and see how results can be improved by using $\tau$

```python
FRBs = GetFRBcat( tau=True )

P1s, zs = LikelihoodRedshift( FRBs['DM'], scenario )
P2s, zs = LikelihoodRedshift( FRBs['DM'], scenario, taus=FRBs['tau'] )
#P2s, zs = LikelihoodRedshift( DMs, scenario, population='sfr', telescope='parkes' )
fig, ax = plt.subplots()
for P1, P2, id, color in zip( P1s, P2s, FRBs['ID'], Rainbow(np.linspace(0,1,len(FRBs))) ):
    PlotLikelihood( zs, P1, measure='z', log=False, ax=ax, color=color, linestyle='--' )
    PlotLikelihood( zs, P2, measure='z', log=False, ax=ax, color=color, linestyle=':', linewidth=3 )
    z_est, z_dev = Likelihood2Expectation( P1, zs, log=False )
    print r"%s redshift(DM) = %.2f +- %.2f" % ( id, z_est, z_dev )
    z_est, z_dev = Likelihood2Expectation( P2, zs, log=False )
    print r"%s redshift(DM,tau) = %.2f +- %.2f" % ( id, z_est, z_dev ) ,'\n'
plt.xlabel('redshift')
plt.show()
```
