```python
def BayesFactorDeviation( P1=0, P2=0, std1=0, std2=0, which_NaN=False ):
    ratio =  P1/P2
    NaN = np.isnan(ratio) + np.isinf(ratio)
    if np.any(NaN):
        print( "%i of %i returned NaN. Ignore in final result" %( np.sum(NaN), len(DMs) ) )
        if which_NaN:
            ix, = np.where( NaN )
            print(ix)

    bayes = np.prod(ratio[~NaN])
    dev = np.sqrt( np.sum( np.array([std1,std2])[:,~NaN]**2 ) )
    return bayes, dev
    
```

```python
def Deviation2Log( x, dev ):
    """ transform 1D linear scaled deviation to 2D log-scaled deviation use in plt.errorbar """
    x_log = np.log10(x)
    dev_log = np.log10(x+dev) - x_log
    return np.array( [ x - 10.**( x_log - dev_log ), 10**( x_log + dev_log ) - x ] )
    
```
